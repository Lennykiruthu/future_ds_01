```python
import pandas as pd
import typing
from IPython.display import display, Markdown
```

```python
def Explore(path : str, name : str):
    df = pd.read_csv(path)
    display(Markdown(name))
    display(Markdown("---"))
    display(df.head(5))
    display(Markdown("### Data Shape"))
    display(df.shape)    
    display(Markdown("### Data Types"))
    display(df.dtypes)    
    display(Markdown("### Missing Values"))
    display(df.isnull().sum().sort_values(ascending=False))      

    # Compute unique values per column
    unique_values = [df[col].nunique() for col in df.columns]
    # Conver to dataframe for easy viewing
    unique_values_df = pd.DataFrame({'Columns' : df.columns, 'Unique Values Count': unique_values})

    display(Markdown("### Unique Values"))
    display(unique_values_df.sort_values(by=unique_values_df.columns[1], ascending=False).head(len(unique_values_df)))
    display(Markdown("### Numeric Data Types Descriptive Statistics"))
    display(df.describe())

Explore(r"D:\local-git-repos\brazillian-ecommerce-analysis\olist_customers_dataset.csv", "## Olist Customers Dataset")
```

```python
# Load CSVs
products = pd.read_csv(r"D:\local-git-repos\brazillian-ecommerce-analysis\olist_products_dataset.csv")
translations = pd.read_csv(r"D:\local-git-repos\brazillian-ecommerce-analysis\product_category_name_translation.csv")


# Get unique category names from each
product_categories = set(products["product_category_name"].dropna().unique())
translated_categories = set(translations["product_category_name"].dropna().unique())

# Find missing ones
missing_categories = product_categories - translated_categories

print(missing_categories)
```
